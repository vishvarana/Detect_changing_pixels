```python
import cv2
import os
import glob
videos=glob.glob('C:/Users/Yuvraj/Desktop/Vishva')
np_video_list=[]
for video in videos: 
    for v in glob.glob(video+'/*.mp4'): 
        np_video_list.append(v) 
i=1
frame_no=1
images = []
for j in range(0,len(np_video_list)): 
    video_data=np_video_list[j]
    cap=cv2.VideoCapture(video_data) 
    
    while cap.isOpened():
        ret,frame=cap.read()
        if ret:
            i = i + 1
            if i % 256 == 0 & frame_no<=200:
                if frame_no==1:
                    name = 'C:/Users/Yuvraj/Desktop/Vishva/ImageData' + '/' +'key_f' + str(frame_no) + '.jpg'
                    cv2.imwrite(name, frame)
                    print('Creating ' + name)
                    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
                    images.append(gray)
                    frame_no+=1
                elif frame_no % 50 !=0 :
                    if frame_no>1 & frame_no<50:
                        name = 'C:/Users/Yuvraj/Desktop/Vishva/ImageData' + '/' +'non_key' + str(frame_no) + '.jpg'
                        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
                        result = cv2.bitwise_and(gray,images[0])
                        cv2.imwrite(name, result)
                        print('Creating ' + name)
                        frame_no+=1
                    elif frame_no>50 & frame_no<100:
                        name = 'C:/Users/Yuvraj/Desktop/Vishva/ImageData' + '/' +'non_key' + str(frame_no) + '.jpg'
                        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
                        result = cv2.bitwise_and(gray,images[1])
                        cv2.imwrite(name, result)
                        print('Creating ' + name)
                        frame_no+=1
                    elif frame_no>100 & frame_no<150:
                        name = 'C:/Users/Yuvraj/Desktop/Vishva/ImageData' + '/' +'non_key' + str(frame_no) + '.jpg'
                        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
                        result = cv2.bitwise_and(gray,images[2])
                        cv2.imwrite(name, result)
                        print('Creating ' + name)
                        frame_no+=1
                    else :
                        name = 'C:/Users/Yuvraj/Desktop/Vishva/ImageData' + '/' +'non_key' + str(frame_no) + '.jpg'
                        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
                        result = cv2.bitwise_and(gray,images[3])
                        cv2.imwrite(name, result)
                        print('Creating ' + name)
                        frame_no+=1
                else :
                    name = 'C:/Users/Yuvraj/Desktop/Vishva/ImageData' + '/' +'key_f' + str(frame_no) + '.jpg'
                    cv2.imwrite(name, frame)
                    print('Creating ' + name)
                    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
                    images.append(gray)
                    frame_no+=1
        else:
            break
    cap.release()
```
