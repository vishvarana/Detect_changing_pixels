```python
import numpy as np
import matplotlib.pyplot as plt
from skimage.io import imread, imshow
from skimage import transform
import cv2
```

```python
vidcap = cv2.VideoCapture('M6 Motorway Traffic.mp4')
def getFrame(sec):
    vidcap.set(cv2.CAP_PROP_POS_MSEC,sec*1000)
    hasFrames,image = vidcap.read()
    if hasFrames:
        if count==1:
            name = "C:/Users/Yuvraj/Desktop/Vishva" + "/" + "traffic_key_f" + str(count) + ".jpg"
            cv2.imwrite(name, image) 
            print("Created "+name)
        elif count % 50 != 0:
            name = "C:/Users/Yuvraj/Desktop/Vishva" + "/" + "traffic_non_key" + str(count) + ".jpg"
            cv2.imwrite(name, image) 
            print("Created "+name)
        else :
            name = "C:/Users/Yuvraj/Desktop/Vishva" + "/" + "traffic_key_f" + str(count) + ".jpg"
            cv2.imwrite(name, image) 
            print("Created "+name)
    return hasFrames
sec = 0
frameRate = 5 #capture an image every 5 seconds
count = 1
success = getFrame(sec)
while success:
    count = count + 1
    sec = sec + frameRate
    sec = round(sec, 2)
    success = getFrame(sec)
```

```python
area_of_interest = [(0, 660),
                    (2000, 660),
                    (800, 350),
                    (1150, 350)]

area_of_projection = [(50, 710),
                      (2050, 710),
                      (850, 400),
                      (1200, 400)]
```

```python
def project_planes(image, src=area_of_interest, 
                   dst=area_of_projection):
    new_image = image.copy() 
    projection = np.zeros_like(new_image)
    tform = transform.estimate_transform('projective', np.array(src), 
                                         np.array(dst))
    transformed = transform.warp(image, tform.inverse)
    return transformed
```

```python
plt.figure(figsize=(12,12))
diff = (project_planes(images[15]) - 
        project_planes(images[0]))
imshow(diff)
```

```python
plt.figure(figsize=(12,12))
absdiff = abs(diff)
imshow(absdiff)
```
